```python
import numpy as np
import pandas as pd
import pathlib
from sklearn.cluster import KMeans
from google.colab import drive
drive.mount('/content/drive')
#for using google drive
```

```python
df=pd.read_csv('/content/drive/MyDrive/testKmeans/DiffList.csv')#check point
dff=df
dff
```

```python
dff=dff.drop(columns=['id', 'imagePath'])
dff
```

```python
model = KMeans(n_clusters=3)
model
```

```python
model.fit(dff)
```

```python
belongToCluster=model.predict(dff)
```

```python
blCtDf = pd.DataFrame(belongToCluster, columns=['ClusterNumber'])
```

```python
dataClusterDF = pd.concat([blCtDf, dff], axis=1, join='inner')
dataClusterDF
```

```python
dataClusterDF.groupby(['ClusterNumber'])['diff'].max()
```

```python
dataClusterDF.groupby(['ClusterNumber'])['diff'].min()
```

```python
df2=pd.read_csv('/content/drive/MyDrive/testKmeans/AVGDiffList.csv')#check point
dff2=df2
dff2
```

```python
dff2=dff2.drop(columns=['id', 'imagePath'])
dff2
```

```python
model2 = KMeans(n_clusters=3)
model2

model2.fit(dff2)

belongToCluster2=model2.predict(dff2)

blCtDf2 = pd.DataFrame(belongToCluster2, columns=['ClusterNumber'])

dataClusterDF2 = pd.concat([blCtDf2, dff2], axis=1, join='inner')
dataClusterDF2
```

```python
dataClusterDF2.groupby(['ClusterNumber'])['AVGDiff'].max()
```

```python
dataClusterDF2.groupby(['ClusterNumber'])['AVGDiff'].min()
```
